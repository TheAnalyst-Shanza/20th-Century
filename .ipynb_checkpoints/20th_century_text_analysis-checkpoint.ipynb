```python
DATA_TXT = "key_events_20th_century.txt"  # from your Exercise 1.4
assert Path(DATA_TXT).exists(), f"Can't find {DATA_TXT} in {os.getcwd()}"

# Read and join into a single string
with open(DATA_TXT, "r", encoding="utf-8") as f:
    lines = [ln.strip() for ln in f if ln.strip()]
text = " ".join(lines)

len(text), text[:300]
```

```python
# Tokenize via TextBlob (uses NLTK under the hood)
blob = TextBlob(text)
tokens = [t.lower() for t in blob.words]

# Frequency
counter_all = collections.Counter(tokens)
top10_all = counter_all.most_common(10)
top10_all
```

```python
# Bar chart: Top 10 words (raw)
sns.set_theme()
plt.figure(figsize=(8,4))
words, counts = zip(*top10_all)
sns.barplot(x=list(words), y=list(counts))
plt.title("Top 10 Most Common Words (raw)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
from nltk.corpus import stopwords
stop = set(stopwords.words("english"))
punct = set(string.punctuation)

clean_tokens = [w for w in tokens if w not in stop and w not in punct and not w.isdigit()]
counter_clean = collections.Counter(clean_tokens)
top10_clean = counter_clean.most_common(10)
top10_clean
```

```python
plt.figure(figsize=(8,4))
words, counts = zip(*top10_clean)
sns.barplot(x=list(words), y=list(counts))
plt.title("Top 10 Most Common Words (stopwords & punctuation removed)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

After removing stopwords and punctuation, high-frequency but uninformative words disappeared (e.g., “the”, “and”), revealing more content-bearing terms. This changes the distribution to emphasize topic-specific vocabulary, which is more useful for downstream analysis and building networks.

```python
# Get (word, POS) tags. TextBlob returns Penn Treebank tags (e.g., NN, NNS, VB, JJ, etc.)
tags = blob.tags  # list of tuples: [(word, pos), ...]

# Count POS labels only
pos_counts = collections.Counter([pos for _, pos in tags])
top10_pos = pos_counts.most_common(10)
top10_pos
```

```python
plt.figure(figsize=(8,4))
labels, counts = zip(*top10_pos)
sns.barplot(x=list(labels), y=list(counts))
plt.title("Top 10 POS Tags (All Tokens)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
clean_set = set(clean_tokens)
clean_pos_counts = collections.Counter([pos for w, pos in tags if w.lower() in clean_set])
clean_pos_counts.most_common(10)
```

```python
# Build per-class frequency on cleaned tokens
noun_tags = {"NN","NNS","NNP","NNPS"}
verb_tags = {"VB","VBD","VBG","VBN","VBP","VBZ"}
adj_tags  = {"JJ","JJR","JJS"}

# word -> pos on first occurrence (simple heuristic)
word_pos = {}
for w, pos in tags:
    wl = w.lower()
    if wl not in word_pos:
        word_pos[wl] = pos

def top_terms_by_pos(target_tags, topn=15):
    # Keep only words of these POS and in cleaned list
    filtered = [w for w in clean_tokens if word_pos.get(w) in target_tags]
    return collections.Counter(filtered).most_common(topn)

top_nouns = top_terms_by_pos(noun_tags, 15)
top_verbs = top_terms_by_pos(verb_tags, 15)
top_adjs  = top_terms_by_pos(adj_tags, 15)
top_nouns[:5], top_verbs[:5], top_adjs[:5]
```

```python
def plot_top(title, items):
    plt.figure(figsize=(8,4))
    if not items:
        plt.title(title + " (no items)")
        plt.show()
        return
    words, counts = zip(*items)
    sns.barplot(x=list(words), y=list(counts))
    plt.title(title)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

plot_top("Top 15 Nouns", top_nouns)
plot_top("Top 15 Verbs", top_verbs)
plot_top("Top 15 Adjectives", top_adjs)
```

Nouns highlight key actors/places/things; verbs reflect actions prevalent in the narrative; adjectives can show sentiment or descriptors. What stands out here (e.g., specific country names, conflict terms, technology terms) likely mirrors the historical emphasis of the source text.

```python
try:
    import pycountry
    country_names = sorted([c.name for c in pycountry.countries])
except Exception:
    # Fallback: minimal list
    country_names = [
        "United States", "USA", "Canada", "Mexico", "United Kingdom", "Britain", "England",
        "France", "Germany", "Italy", "Spain", "Russia", "Soviet Union", "China", "Japan",
        "India", "Pakistan", "Australia", "New Zealand", "South Africa", "Brazil", "Argentina",
        "Turkey", "Iran", "Iraq", "Israel", "Egypt", "Saudi Arabia", "Korea", "Vietnam"
    ]

text_lc = text.lower()
counts = {}
for name in country_names:
    # naive count; tweak with regex for whole-word matches
    pattern = r"\b" + re.escape(name.lower()) + r"\b"
    counts[name] = len(re.findall(pattern, text_lc))

# Keep only mentioned ones
country_freq = {k:v for k,v in counts.items() if v>0}
country_df = (pd.Series(country_freq)
              .sort_values(ascending=False)
              .reset_index()
              .rename(columns={"index":"country", 0:"count"}))
country_df.head(20)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=country_df.head(20), x="country", y="count")
plt.title("Top Country Mentions (Top 20)")
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

The frequency distribution shows which nations dominate the text. Spikes can reflect geopolitical focus (e.g., world wars, decolonization, Cold War). Note that name variants (e.g., “United States” vs “USA”) were included to improve matching.
